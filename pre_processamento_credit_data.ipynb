# Tecnicas para o tratamento de dados inconsistentes

 * Apagar as colunas (NÃ£o recomendado pois vai perder os dados)
 * Apagar somente os registros com probelmas
 * Preencher os dados manualmente (Ã‰ melhor dos mundos mas Ã© inviabel)
 * Preencher os valores com a mÃ©dia

```python
#Apagar as colunas
base.drop('age',1,inplace=True)
base
```

```python
#Apagar somente os registros com probelmas
base.drop(base[base.age<0].index,inplace = True)
base
```

```python
#Preencher os valores com a mÃ©dia
base['age'][base.age>0].mean()
```

```python
base.loc[base['age']<0] = base['age'][base.age>0].mean()
```
